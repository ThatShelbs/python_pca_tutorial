### Variance Inflation Factor (VIF)

Variance Inflation Factor (VIF) is a measure of how much the variance of a regression coefficient is inflated due to collinearity with other variables. Here's a breakdown of what you need to know about VIF and how to interpret it:


#### Purpose of VIF

The main purpose of calculating VIF is to detect multicollinearity among the predictor variables in a regression model. Multicollinearity can cause problems in multiple regression models by making it difficult to determine the effect of each predictor.


#### Interpretation of VIF


VIF = 1: No correlation between this predictor and any other predictor.

1 < VIF < 5: Moderate correlation but generally acceptable.

VIF > 5: Potentially problematic level of multicollinearity. Some sources suggest that a VIF above 10 is a sign that the coefficient estimates are poorly estimated due to multicollinearity.

VIF > 10: Very high correlation, problematic multicollinearity. You may consider removing this variable from your model.

#### Implications of Extremely High VIF


Unstable Estimates: The estimates of coefficients will be highly unstable. Small changes in the data can lead to dramatic changes in the model's coefficients.

Inflated Standard Errors: The standard errors of the coefficients will be large, making hypothesis tests like t-tests unreliable.

Interpretation Issues: It will be difficult to determine the individual effect of correlated predictors on the response variable.

## Kaiser-Meyer-Olkin (KMO) test

The Kaiser-Meyer-Olkin (KMO) test is a measure of how suitable your data is for Factor Analysis and Principal Component Analysis (PCA). The KMO measure assesses the degree of common variance among variables, i.e., the proportion of variance that might be common variance among variables. Higher values indicate that Factor Analysis is likely appropriate.

#### Purpose of KMO Test

The purpose of the KMO test is to determine the adequacy of conducting a factor analysis on a given dataset. It helps answer the question: "Do the variables share enough common variance to justify the use of factor analysis?"

#### KMO Calculation

The KMO measure ranges from 0 to 1, and is calculated for individual variables as well as for the overall dataset.

#### Interpretation of KMO Values:

- KMO values close to 1: Indicate that the variables have a high level of common variance, and factor analysis is likely useful.

- KMO values between 0.8 and 1: Considered excellent.

- KMO values between 0.7 and 0.8: Considered good.

- KMO values between 0.6 and 0.7: Considered mediocre.

- KMO values between 0.5 and 0.6: Considered poor.

- KMO values below 0.5: Indicate that factor analysis is likely inappropriate

### VIF vs KMO 

Local Collinearity vs. Common Variance: VIF and KMO measure different properties. A high VIF indicates that a specific predictor is strongly collinear with one or more other predictors, while a low KMO indicates that the feature does not have much common variance with the rest of the variables in the dataset as a whole. The feature with high VIF might be highly collinear with a subset of variables but not share common variance broadly across the dataset.

### Scree Plot

A scree plot in factor analysis or principal component analysis displays eigenvalues of factors or components in descending order. It helps determine the optimal number of factors to retain. The plot typically shows a sharp drop (the "elbow") followed by a more gradual decline. Factors with eigenvalues less than 1 are often considered less significant, as they explain less variance than a single standardized variable.

```python
# Perform Factor Analysis with varimax rotation
fa = FactorAnalyzer(rotation='varimax', n_factors=df.shape[1])
fa.fit(df)

# Eigenvalues (variance explained by each factor before rotation)
ev, v = fa.get_eigenvalues()

num_factors = np.sum(ev >= 1)

# Plot the scree plot
plt.figure(figsize=(12, 8))
plt.plot(range(1, len(ev)+1), ev, 'o-', color='b', markersize=8)
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot')
plt.axhline(y=1, color='r', linestyle='--')
plt.grid()
plt.show()

print("Number of factors above threshold: {}".format(num_factors))
```

```python
# Perform Factor Analysis with varimax rotation
fa = FactorAnalyzer(rotation='varimax', n_factors=num_factors)
fa.fit(df)

# Create a DataFrame with loadings of each feature across the different factors
loadings = pd.DataFrame(fa.loadings_, index=df.columns, columns=[f'Factor{i+1}' for i in range(num_factors)])

# Display complete loadings table without applying the threshold for initial review
print("Loadings (without threshold):")
print(loadings)

# Apply a threshold (e.g., R square >= 0.40) to the loadings
threshold = 0.40
thresholded_loadings = loadings.copy()
thresholded_loadings[(loadings**2) < threshold] = None

# Print loadings table with applied threshold
print(f"\nLoadings (Features with R^2 >= {threshold}):")
print(thresholded_loadings)

# Calculate and print the report of variance explained by each factor
variance = fa.get_factor_variance()
variance_explained_table = pd.DataFrame({
    'Eigenvalue': variance[0],
    'Proportion Variance': variance[1],
    'Cumulative Variance': variance[2]
}, index=[f'Factor{i+1}' for i in range(num_factors)])

print("\nVariance Explained by Each Factor:")
print(variance_explained_table)
```

```python
# Create a list of column names for sorting
sort_columns = [f'Factor{i+1}' for i in range(num_factors)]

# Sort the DataFrame
thresholded_loadings.sort_values(by=sort_columns, 
                                 ascending=[False] * num_factors, 
                                 inplace=True)

thresholded_loadings
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(thresholded_loadings, annot=True, cmap='viridis', cbar_kws={'orientation': 'horizontal'})
plt.title(f'Loadings (Features with R^2 >= {threshold} marked, others as NaN)')
plt.show()
```
